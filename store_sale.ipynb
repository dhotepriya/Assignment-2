```python
total=sales_data[['Jan','Feb','Mar','Apr']].sum() # sum of sales from jan-apr
total
```

```python
avg=sales_data[['Jan','Feb','Mar','Apr']].mean() # avg of sales from jan-apr
avg
```

```python
Year= sales_data[['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'July', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']].sum(axis=1) # year sales of all stores--> axis=1--cols
Year
```

```python
month = sales_data.groupby('city')[['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'July', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']].sum() # each month total sales of cities
month
```

```python
avg_month = sales_data.groupby('city')[['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun','July', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']].mean() # ecah month avg sales of cities 
avg_month
```

```python
avg_month['Overall_Avg'] = avg_month.mean(axis=1)

# City with the highest average sales
highest_city = avg_month['Overall_Avg'].idxmax()
highest_value = avg_month['Overall_Avg'].max()

print(f"City with the highest average sales: {highest_city} ({highest_value})")
```
